```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('/content/Race and Ethnicity Projections by The Year 2060.xlsx')
df.head()
```

```python
Year = df['Year']
```

```python
White = df['White*']
```

```python
Black = df['Black']
```

```python
Hispanic = df['Hispanic']
```

```python
Asian = df['Asian']
```

```python
Multiple_Races = df['Multiple Races']
```

```python
Other = df['Other**']
```

```python
plt.style.use('fivethirtyeight')
plt.bar( Year, White, width = .75, label= 'White'); 
plt.bar( Year + .75, Hispanic, width = .75, label= 'Hispanic'); 
plt.bar( Year + 1.5, Black, width = .75, label= 'Black');
plt.bar( Year + 2.25, Asian, width = .75, label= 'Asian');
plt.bar( Year + 3, Multiple_Races, width = .75, label= 'Multiple Races');
plt.bar( Year + 3.75, Other, width = .75, label= 'Other');
plt.xticks(fontsize = 15);
plt.yticks(fontsize = 15);
plt.xlabel('Year', fontsize = 18);
plt.ylabel('Percentage', fontsize = 18);
plt.title('Race and Ethnicity Projections (2020-2060)', fontsize = 20);
```

```python
Races_And_Ethnicity2020 = [White[0], Black[0], Hispanic[0], Asian[0], Multiple_Races[0] + Other[0]]
```

```python
Labels_Races_And_Ethnicity2020 = ['White' , 'Black', 'Hispanic', 'Asian', 'Multiple Races & Other Combined']
```

```python
plt.style.use('tableau-colorblind10')
plt.pie(Races_And_Ethnicity2020,explode=[0,.2,0,0,0,] ,
        autopct='%1.1f%%',labels= Labels_Races_And_Ethnicity2020, 
        wedgeprops= {'edgecolor' : 'black'}, shadow = True);
plt.title('The Percentage of Black Race In The US Population (2020) ', fontsize = 16);
```

```python
plt.style.use('seaborn')
plt.plot(Year, White, color ='m', linewidth = 5, markersize = 30, marker= '.' );
plt.xlabel('Year', fontsize = 15);
plt.ylabel('Percentage', fontsize = 15);
plt.title('Projection of White Race In The US Population', fontsize = 20);
```

```python
plt.style.use('seaborn-bright');
plt.plot(Year, Other, label= 'Other');
plt.plot(Year, Multiple_Races, label= 'Multiple Races');
plt.legend(loc= (1.01, .95))
plt.grid(c = 'k', alpha = .2)
plt.xlabel('Year', fontsize = 15);
plt.ylabel('Percentage', fontsize = 15);
plt.title('Other vs Multiple Races Projection (2020-2060)', fontsize = 20);
```
